```python
from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ["Medicare", 'Crime', 'Immigration', 'Unemployment', 'Social Security']
pytrends.build_payload(kw_list, cat=0, timeframe='today 1-m', geo = 'US-MN', gprop='')
```

```python
topics_by_region = pytrends.interest_by_region(resolution='CITY', inc_geo_code=False, inc_low_vol=True)
```

```python
topics_by_region.head(6)
```

```python
import pandas as pd
topics_by_region['region'] = ['Duluth', 'Fargo', 'La Crosse', 'Mankato', 'Minneapolis', 'Rochester']
```

```python
topics_by_region = topics_by_region.reset_index()
```

```python
del topics_by_region['geoName']
topics_by_region.head(6)
```

```python
topics_by_region2 = topics_by_region.melt(id_vars = ['region'],
                                    value_vars = ['Medicare', 'Crime', 'Immigration', 'Unemployment', 'Social Security'])
```

```python
topics_by_region2 = topics_by_region2.rename(columns = {'variable':'topic', 'value':'popularity'})
```

```python
topics_by_region2.head(30)
```

```python
from google.cloud import storage
client = storage.Client()
print('default project: {}'.format(client.project))
```

```python
bucket = client.bucket('msba-trends')
```

```python
topics_by_region2.to_csv('mn_topics_by_region.csv', index = False)
blob = bucket.blob('mn_topics_by_region.csv')
blob.upload_from_filename('mn_topics_by_region.csv')
```

```python
topics_by_day = pytrends.interest_over_time()
del topics_by_day['isPartial']
```

```python
topics_by_day = topics_by_day.reset_index()
topics_by_day.head(29)
```

```python
topics_by_day.dtypes
```

```python
topics_by_day.to_csv('mn_topics_by_day.csv', index = False)
blob2 = bucket.blob('mn_topics_by_day.csv')
blob2.upload_from_filename('mn_topics_by_day.csv')
```
