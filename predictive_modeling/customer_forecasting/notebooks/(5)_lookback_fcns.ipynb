look back 6 - 9 weeks, avg visitors for that weekday(Sun - Sat), impute the first month of being open avg visitors value for when stores r closed or have been open for only 6 weeks (functions by store, genre, and area)

```
def lookback_store(df):
    i = 0
    values = []
    
    for i in range(len(df)):
        impute_val = 0
        if (df['closed_flag'][i] == 0) | (df['closed_flag'][i] == 1):
            
            impute_val = df[(df['air_store_id'][i] == df['air_store_id']) &
            ((df['calendar_date'] == df['calendar_date'][i] - pd.Timedelta(weeks = 6)) |
            (df['calendar_date'] == df['calendar_date'][i] - pd.Timedelta(weeks = 7)) |
            (df['calendar_date'] == df['calendar_date'][i] - pd.Timedelta(weeks = 8)) |
            (df['calendar_date'] == df['calendar_date'][i] - pd.Timedelta(weeks = 9)))]['visitors'].dropna().mean()
                
            if str(impute_val) != 'nan':
                values.append(impute_val)
                
            else:
                impute_val = df[(df['air_store_id'][i] == df['air_store_id']) &
                                ((df['month'] == df['earlist_open_date'][i].month) &
                                 (df['year'] == df['earlist_open_date'][i].year))]['visitors'].dropna().mean()
                
                values.append(impute_val)
            
        else:
            impute_val = df[(df['air_store_id'][i] == df['air_store_id']) &
                            ((df['month'] == df['earlist_open_date'][i].month) &
                             (df['year'] == df['earlist_open_date'][i].year))]['visitors'].dropna().mean()
            
            values.append(impute_val)
        
        i += 1
    
    df['lookback_visitors_avg_store'] = values
            
    return df
```
