```python
import pandas as pd
import numpy as np
import warnings  
from fancyimpute import KNN

warnings.filterwarnings('ignore')
```

```python
master = pd.read_csv("../input/final.csv")
```

```python
master = master.drop(['solar_radiation', 'deepest_snowfall', 'total_snowfall', 'avg_vapor_pressure', 'avg_local_pressure', 'avg_sea_pressure', 'cloud_cover'], axis=1)
```

```python
dates = master['calendar_date'].unique()
```

```python
master = master.sort_values(['calendar_date', 'air_store_id'])
master.head(10)
```

```python
initial_summary = master.describe()
```

```python
# avg_temperature
```

```python
final = pd.DataFrame()

cols_of_interest = ['avg_temperature', 'latitude', 'longitude']

for date in dates:
    temp = master[master['calendar_date'] == date]
    temp2 = temp[cols_of_interest]
    temp2 = KNN(k=3).fit_transform(temp2)
    temp['avg_temperature'] = temp2[:,0]
    final = final.append(temp)

master = final
```
